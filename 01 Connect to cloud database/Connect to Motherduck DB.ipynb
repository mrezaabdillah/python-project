## Step 4: Connecting to Mother Duck
You have provided information about the database name (my_db) and the Mother Duck token. Now, establish a connection to Mother Duck using DuckDB and include the database name and token as parameters.

```python
database_name = 'your database name'
motherduck_token = 'your token connection'

# Buat koneksi ke Motherduck dengan menyertakan nama database dan token
con = duckdb.connect(f'md:{database_name}?motherduck_token={motherduck_token}&saas_mode=true')
```

## Step 5: Executing SQL Query and Displaying Results

```python
query = "SELECT * FROM city_economic LIMIT 1"
df = con.execute(query).fetch_df()
```

```python
df
```

## Step 6: Processing the 'geometry' Column containing WKB Blob into a GeoDataFrame

```python
# Misal, asumsikan bahwa kolom 'geometry' berisi blob WKB
df['geometry'] = df['geometry'].apply(lambda x: loads(bytes(x)))
```

```python
# Sekarang, Anda dapat membuat GeoDataFrame
gdf = gpd.GeoDataFrame(df, geometry='geometry', crs= 'EPSG:4326')
```

```python
# Set CRS ke EPSG 4326
gdf = gdf.set_crs(epsg=4326)
```
