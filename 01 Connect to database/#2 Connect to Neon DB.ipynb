### Step 2: Import Library

```python
import sqlalchemy
import pandas as pd
import geopandas as gpd
from shapely import wkb
from shapely.wkb import loads
```

### Step 3: Connecting to Neon DB with sqlalchemy
Replace 'database_connection_string' with the actual connection details to your Neon DB. The engine is essential for connecting and interacting with the database.

```python
# Koneksi ke database PostgreSQL
engine = sqlalchemy.create_engine("your connection string")
```

### Step 4: Querying and Reading Data

```python
# Query SQL untuk mengambil data dari tabel tertentu
query = "SELECT * FROM streamlit.city_economic LIMIT 1"

# Gunakan metode read_sql_query dari pandas untuk membaca hasil query ke dataframe
df = pd.read_sql_query(query, engine)
```

```python
df
```

### Step 5: Processing the 'geometry' Column containing WKB Hex to GeoDataFrame

```python
df['geometry'] = df['geom'].apply(lambda x: wkb.loads(bytes.fromhex(x)))
```

```python
# Sekarang, Anda dapat membuat GeoDataFrame
gdf = gpd.GeoDataFrame(df, geometry='geometry', crs= 'EPSG:4326')
```

```python
# Set CRS ke EPSG 4326
gdf = gdf.set_crs(epsg=4326)
```
